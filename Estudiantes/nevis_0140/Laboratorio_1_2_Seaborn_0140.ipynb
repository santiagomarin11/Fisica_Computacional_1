```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Hacer el analisis estadístico de las variables: Millas por galón (mpg) vs potencia (horsepower) in the dataset: millas_por_galón.csv. Contruya el fit de al 95 C.L.

```python
df = pd.read_csv("/content/drive/MyDrive/Computacional-1/millas_por_galon.csv")
df
```

**Análisis estadístico básico:**

```python
sns.scatterplot(data=df, x ="mpg",y="horsepower", color="navy")
plt.grid()
plt.title("Millas por galón vs Potencia")
```

```python
#Grafico de dispersión y el ajuste lineal con el 95% de confianza
sns.regplot(data = df, x="mpg", y ="horsepower",ci=95,color ="navy")
plt.grid()
plt.title("Tendencia entre las variables $mpg$ y horsepower")
```

```python
#Correlación entre mpg y horsepower
corr = df["mpg"].corr(df["horsepower"])
print("El coeficiente de correlación entre mpg y horsepower es:",corr)
```

```python
#mapa de calor de mpg y horsepower.
sns.heatmap(df[["mpg","horsepower"]].corr(),annot=True,cmap="coolwarm")
```

A partir de las graficas anteriores se encontro un coeficiente de correlación del $-0.78$, lo cual indica una relación negativa fuerte entre ambas variables, esto significa que a medida que aumenta la potencia del motor, la eficiencia del combustible disminuye significativamente, el mapa de calor y el análisis de correlación confirman una dependencia inversa clara, aunque no es perfecta, lo cual indica que pueden existir otros factores que también influyen al consumo del combustible
