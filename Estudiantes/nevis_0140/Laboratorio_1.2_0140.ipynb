```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("penguins").dropna().values
penguins = pd.DataFrame(data, columns=["especie", "isla", "longitud_pico (mm)", "profundidad_pico (mm)", "longitud_aleta (mm)", "masa_corporal (g)", "genero"])

penguins
```

##1. ¿Cuáles son las profundidades del pico máxima y mínima de los pinguinos de la isla Torgersen?

```python
torgersen = penguins[penguins["isla"] == "Torgersen"]

min_depth = torgersen["profundidad_pico (mm)"].min()
max_depth = torgersen["profundidad_pico (mm)"].max()

print("Profundidad mínima:", min_depth)
print("Profundidad máxima:", max_depth)
```

## 2. ¿De qué isla, especie y género es el pingüino con la mayor masa corporal?

```python
max_mass = penguins.loc[penguins["masa_corporal (g)"].idxmax()]
print(max_mass[["isla","especie","genero","masa_corporal (g)"]])
```

## 3. ¿Cuál es la masa corporal media de los pinguinos macho que no son de la especie Gentoo?

```python
male_not_gentoo = penguins[(penguins["genero"] == "Male") & (penguins["especie"] != "Gentoo")]
mean_mass = male_not_gentoo["masa_corporal (g)"].mean()
print("La masa corporal media de los pinguinos macho que no son de la especie Gentoo es", mean_mass,"g")
```

##4. Grafique la distribución de la longitud del pico. Añada el valor medio y la desviación estándar en el título de la figura

```python
mean_val = penguins["longitud_pico (mm)"].mean()
std_val = penguins["longitud_pico (mm)"].std()

print(mean_val , std_val)
```

```python
penguins['longitud_pico (mm)'].hist(bins=25, color ="cornflowerblue")
plt.title(f"distribución de la longitud del pico \nMedia = {mean_val:.2f}, SD = {std_val:.2f}")
plt.xlabel("Longitud del pico $mm$")
plt.ylabel("Frecuencia")
plt.show()
```
