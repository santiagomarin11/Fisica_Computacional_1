```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```


* Hacer el analisis estadístico de las variables: Millas por galón (mpg) vs potencia (horsepower) in the dataset: millas_por_galón.csv. Contruya el fit de al $95$ C.L.     



  



```python
data = "https://raw.githubusercontent.com/anferivera/Fisica_Computacional_1/refs/heads/main/Sesiones/seaborn/millas_por_galon.csv"
mxg =  pd.read_csv(data)
mxg.head()
```

```python
g = sns.relplot(
    data=mxg,
    x="horsepower",
    y="mpg",
    hue="origin",
    size="weight",
    sizes=(40, 400),
    alpha=0.5,
    palette="viridis"
)

#ajuste lineal
sns.regplot(
    data=mxg,
    x="horsepower",
    y="mpg",
    scatter=False,
    ax=g.ax,
    color="purple"
)
```
