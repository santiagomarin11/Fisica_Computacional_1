```
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Dataframe de datos acerca de diferentes especies de pinguinos:

```
data = sns.load_dataset("penguins").dropna().values
df = pd.DataFrame(data, columns=["especie", "isla", "longitud_pico (mm)", "profundidad_pico (mm)", "longitud_aleta (mm)", "masa_corporal (g)", "genero"])
df.head()
```


1.   ¿Cuáles son las profundidades del pico máxima y mínima de los pinguinos de la isla Torgersen?

```
print(f"Las profundidades del pico máxima y mínima de los pinguinos de la isla Torgersen son respectivamente { df["profundidad_pico (mm)"].max()
}mm y {df["profundidad_pico (mm)"].min()}mm")
```

2.    ¿De qué isla, especie y género es el pingüino con la mayor masa corporal?
       **Ayuda:** utilice la función `argmax()`

```
fila = df["masa_corporal (g)"].idxmax()

print(f"""El pinguino con mayor masa corporal correspondiente a {df.loc[fila, "masa_corporal (g)"]} g
es de la isla {df.loc[fila, "isla"]} ,
de especie {df.loc[fila, "especie"]},
y de genero {df.loc[fila, "genero"]} """)
```

3.   ¿Cuál es la masa corporal media de los pinguinos macho que no son de la especie Gentoo?

```
df_sinGentoo = df[df["especie"] != "Gentoo"]
df2 = df_sinGentoo[df_sinGentoo["genero"] == "Male"]
print(f"La masa corporal media de los pinguinos macho que no son de la especie gentoo es {df2["masa_corporal (g)"].mean()} ")
```

4. Grafique la distribución de la longitud del pico. Añada el valor medio y la desviación estándar en el título de la figura. Ayuda: utilice la funcion hist()

```
media , std = df["longitud_pico (mm)"].mean() , df["longitud_pico (mm)"].std()

plt.hist(df["longitud_pico (mm)"] , bins = 20 , color = "pink")
plt.xlabel('longitud_pico (mm)')
plt.ylabel('N')
plt.axvline(media, linestyle="--" , color = "red" , label = f"valor medio:{round(media,2)}")

plt.axvline(media + std, linestyle=":" , color = "red" , label = f"desviacion estandar:{round(std,2)}")
plt.axvline(media - std, linestyle=":" , color = "red" )
plt.title('Distribución de la longitud del pico')

plt.legend()
plt.grid()
plt.show()
```

5.   ¿Cuáles son las islas y especies presentes en el conjunto de datos? ¿Con qué frecuencia aparecen? Grafique la distribución de la masa corporal para los pingüinos.



```
islas = df["isla"].unique()
especies = df["especie"].unique()

print(f"""Las islas presentes en el conjunto de datos son {str(islas)}
      la frecuencia en la cual aparece cada {df["isla"].value_counts()}""")


print(f"""Las especies presentes en el conjunto de datos son {str(especies)}
       la frecuencia en la cual aparece cada {df["especie"].value_counts()}""")
```

```
media , std = df["masa_corporal (g)"].mean() , df["masa_corporal (g)"].std()

plt.hist(df["masa_corporal (g)"] , bins = 20 , color = "pink")
plt.xlabel('masa_corporal (g)')
plt.ylabel('N')
plt.axvline(media, linestyle="--" , color = "red" , label = f"valor medio:{round(media,2)}")

plt.axvline(media + std, linestyle=":" , color = "red" , label = f"desviacion estandar:{round(std,2)}")
plt.axvline(media - std, linestyle=":" , color = "red" )
plt.title('Distribución de la masa corporal de los pinguinos')

plt.legend()
plt.grid()
plt.show()
```
