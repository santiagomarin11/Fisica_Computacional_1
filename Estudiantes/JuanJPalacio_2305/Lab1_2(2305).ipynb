```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/juanpalacio10/Fisica_Computacional_1/refs/heads/main/Sesiones/seaborn/millas_por_galon.csv'
fmri = pd.read_csv(url)
fmri
```

```python
sns.regplot(fmri['horsepower'],fmri['mpg'],ci=95)
plt.title('Millas por galón (mpg) vs potencia (horsepower) con C.L 95')
```
