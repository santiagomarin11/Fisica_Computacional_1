```python
import librosa
import numpy as np


signal, sr = librosa.load('/content/drive/MyDrive/musica /Sesiones_Fourier_Nota_CM_piano.mp3', sr=None)

print(f"Sample Rate: {sr}")
print(f"Signal Shape: {signal.shape}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.plot(signal)
plt.title("MP3 Signal Waveform")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()
```

```python


# 1. Sample the signal seen before (t,y)
#fs = 100 # Sampling frequency (Hz) define in t

# 2. Compute the Fourier Transform
N = len(signal)
yf = fft(signal)
xf = fftfreq(N, 1 / sr)
```

```python
sr
```

```python
# 3. Plot the magnitude spectrum
mask = (np.log(xf[:N//2]) >= 5) & (np.log(xf[:N//2]) <= 7)
plt.plot(np.log(xf[:N//2])[mask],(1.0/N)*np.abs(yf[:N//2])[mask])
```

```python
from scipy.signal import find_peaks

peaks, _ = find_peaks((1.0/N)*np.abs(yf[:N//2]), height=0.010)
```

```python
plt.plot(
    xf[:N//2][peaks],
    (1.0/N)*np.abs(yf[:N//2])[peaks],
    "ro"
)

plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```
