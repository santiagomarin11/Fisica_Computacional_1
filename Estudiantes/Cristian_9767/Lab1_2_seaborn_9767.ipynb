## Análisis de la Relación entre Millas por Galón (mpg) y Potencia (horsepower)

Este laboratori se encarga de analizar la relación entre las `Millas por galón (mpg)` y la `Potencia (horsepower)` de los coches en el dataset `millas_por_galon.csv`.

Para esto, se usa una gráfica de regresión lineal con `seaborn`, que es una herramienta útil para ver cómo dos variables numéricas se relacionan.

### El Gráfico de Regresión (Regplot)

El gráfico que se genera es un `regplot` y se ve así:

```python
df = pd.read_csv('millas_por_galon.csv')
sns.regplot(x='horsepower', y='mpg', data=df, ci=95, line_kws={'color': 'red'})
plt.title('Millas por galón vs Potencia (con 95% C.L.)')
plt.xlabel('Potencia (horsepower)')
plt.ylabel('Millas por galón (mpg)')
plt.show()
```

**¿Qué estamos viendo aquí?**

*   **La Línea Roja:** Esta es la línea de regresión. Nos da una idea de la tendencia general: a medida que la potencia (horsepower) aumenta, las millas por galón (mpg) tienden a disminuir. Lo cual tiene sentido, un motor más potente consume más.
*   **El Área Sombreada Roja:** Es el **Intervalo de Confianza del 95% (95% C.L.)**. La `C.L.` significa 'Confidence Level' o Nivel de Confianza. Básicamente, esta área nos dice que, si repitiéramos este análisis muchas veces, el 95% de las veces, el **verdadero valor promedio** de `mpg` para un `horsepower` específico caería dentro de esta banda. No es que una nueva línea de regresión de otra muestra caiga completamente ahí, sino que el *promedio real* de mpg para cada nivel de potencia está ahí dentro con un 95% de seguridad.
