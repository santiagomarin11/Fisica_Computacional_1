
## Series de pandas

Una serie de pandas es un arreglo unidimensional de datos indexados

**bold text**![](https://i.imgur.com/lYSmUIx.png)


Podemos crear una serie a partir de una lista o arreglo mediante el constructor `Series`

````
# datos unidimensionales
lista = [3, -1, 7, 6, 22]

# indices
indicesmios = ["C", "B", "C", "D", "E"]

# creacion de la serie
serie = pd.Series(data=lista, index=indicesmios, dtype="int8")
serie
````

Podemos acceder al tipo de dato, valores e índices de la serie con los atributos `dtype`, `values` e `index`, respectivamente


## Dataframe de pandas

La siguiente estructura fundamental en Pandas es el *Dataframe*. Si una serie es un análogo de un arreglo unidimensional con índices flexibles, un *DataFrame* es un análogo de un arreglo bidimensional con índices de fila y columna flexibles

![](https://i.imgur.com/1s3PPpq.png)

````
# Construir un DataFrame a partir de un diccionario

#Declarando el diccionario de nombre "data"
misdatos = {
    "Pais": ["Suiza", "Australia", "Canada","Colombia"],
    "Capital": ["Berna", "Canberra", "Ottawa","Bogota"],
    "Pobla_en_millones": [8.6, 25.7, 38, 51]
}

misindices = ["A", "B","2","3"]

#Creando el DataFrame, al que estoy nombrando como "tabla"
tabla = pd.DataFrame(data = misdatos, index= misindices)
tabla
````

# Example 1: Serie example and Dataframe from a dictionary

````
# creamos un diccionario de tres keys o claves y su value es una lista de 8 elem
randomdict = {
    'america':np.random.rand(8),
    'europa':np.random.rand(8),
    'asia':np.random.rand(8),  # la ultima coma es obligatoria, recordar dict.
    }

#A partir de un diccionario o una lista, se puede declarar una serie (estructura de datos unidimensional)
print('************* Serie *****************')
una_serie = pd.Series(randomdict)
print(una_serie)

print('\n************* Dataframe *************\n')
#A partir de un diccionario o una lista, se puede declarar un Dataframe
mi_df= pd.DataFrame(randomdict)
print(mi_df)
````

## Let's do statistic to the DataFrame

````
type(list(df['longitude'])[0])
````

````
listacolumnas= df.columns.values
print('DataFrames columns:\n',listacolumnas)

print(f"\n COLUMNA \t Mean \t\t Median \t Std")
for name in listacolumnas: # ciclo sobre columnas
  newlist = list(df[name]) # extraer la columna de clave name
  #newlist = df[name].values  otra forma de extraer una columna en su forma de lista

  #antes de hacer promedios verificar si los datos son numericos pues podrian ser strings
  if(type(newlist[0]) == int or type(newlist[0])== float): #solo promedios en int o floats

   x = np.array(newlist) # arreglo lista numpy de int o floats
   print(f"{name:.10s}\t {x.mean():.3f} \t {np.median(x):.3f} \t {x.std():.5}")
   #notar que mean y std son metodos del arreglo , pero median es una funcion de numpy

#controles de impresion una variable usando fstrings
#{var:.10s} 10 caracteres
#{var:6.2f}  6 cifras, 2 decimales
````

* The mean is the numerical average calculated by summing all values and dividing by the count, while the median is the middle value in an ordered dataset.
* The mean is the arithmetic average (sum divided by count), whereas the median is the 50th percentile, or middle point.

## plotly and Matplolib

* Plotly is an open-source Python library used for creating interactive, publication-quality graphs and analytical web applications. It allows users to build a wide range of visualizations, including line plots, scatter plots, bar charts, 3D graphs, statistical charts, and maps, which feature interactive elements like hover tooltips, zooming, and panning.

* Matplolib vs Plotly: The core difference is that Matplotlib excels at creating static, highly customizable plots for academic or print publications, while Plotly is designed for creating modern, interactive, web-based visualizations for dashboards and data exploration.
